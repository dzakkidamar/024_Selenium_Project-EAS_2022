```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,5):
    browser.get(f'https://www.tokopedia.com/search?navsource=home&page={index}&q=baju%20lebaran%20pria%20wanita&source=universe&srp_component_id=02.02.01.01&st=product')
    time.sleep(50)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='css-974ipl')
```

```python
    for row in rows:
        Product_Name = row.find(class_='prd_link-product-name css-3um8ox')
        if Product_Name:
            Product_Name = Product_Name.text
            
        Sold = row.find(class_='prd_label-integrity css-1duhs3e')
        if Sold:
            Sold = Sold.text
            
        Rating = row.find(class_='prd_rating-average-text css-t70v7i')
        if Rating:
            Rating = Rating.text
            
        Discount = row.find(class_='prd_badge-product-discount css-1qtulwh')
        if Discount:
            Discount = Discount.text
        
        Cashback = row.find(class_='prd_label-product-price css-tolj34')
        if Cashback:
            Cashback = Cashback.text
        
        Price = row.find(class_='prd_link-product-price css-1ksb19c')
        if Price:
            Price = Price.text
        
        Seller_Loct = row.find(class_='prd_link-shop-loc css-1kdc32b flip')
        if Seller_Loct:
            Seller_Loct = Seller_Loct.text
            
        
            
        #Nama_Toko = row.find(class_='prd_link-shop-name css-1kdc32b flip').text
        #Lokasi_Penjual = row.find(class_='prd_link-shop-loc css-1kdc32b flip').text
        
        
        df = {
            'Product Name': Product_Name,
            'Terjual': Sold,
            'Rating': Rating,
            'Dapatkan Diskon!': Discount,
            'Dapatkan Cashback': Cashback,
            'Harga Sekarang!': Price,
            'Lokasi_Penjual': Seller_Loct
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
